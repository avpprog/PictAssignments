```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Amazon Books Data.csv")
df.head()
```

```python
df = df.loc[:, df.columns[~df.columns.str.contains('^Unnamed')]]
print("Columns after removal of unnecessary ones:")
print(df.columns)
df.head()
```

```python
df_cleaned = df.dropna(how='all')
print(df_cleaned)
```

```python
df.shape
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values)
```

```python
# Get descriptive statistics for numerical columns
statistics = df.describe()
print("Initial Statistics:\n", statistics)
```

```python
# Check data types and dimensions of the dataset
print("\nData Types:\n", df.dtypes)
print("\nData Dimensions:", df.shape)
```

```python
# Summarize the types of variables
print("\nVariable Data Types:\n", df.dtypes)
```

```python
# Convert rating and review_count to numeric ,'edition'
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  # Converts to float
df['review_count'] = pd.to_numeric(df['review_count'], errors='coerce')  # Converts to float
df['review_count'] = df['review_count'].fillna(0).astype('int64')  # Fill NaN with 0 and convert to int64
df['edition'] = pd.to_numeric(df['review_count'], errors='coerce')  # Converts to float
df['edition'] = df['review_count'].fillna(0).astype('int64')  # Fill NaN with 0 and convert to int64

print(df.dtypes)
```

```python
# Convert price to numeric by removing currency symbols
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True)
df['price'] = pd.to_numeric(df['price'], errors='coerce')  # Converts to float
print(df.dtypes)
```

```python
# Convert publish_dat to datetime
df['publish_date'] = pd.to_datetime(df['publish_date'], errors='coerce')
print(df.dtypes)
```

```python
# Importing SimpleImputer from sklearn for handling missing values
from sklearn.impute import SimpleImputer

# Handle missing values in numerical columns using the median
numerical_columns = ['rating', 'review_count', 'price','edition']
imputer_numerical = SimpleImputer(strategy='median')
df[numerical_columns] = imputer_numerical.fit_transform(df[numerical_columns])

# Handle missing values in categorical columns using the mode (most frequent value)
categorical_columns = ['title', 'description', 'author', 'isbn10','isbn13','best_seller']
imputer_categorical = SimpleImputer(strategy='most_frequent')
df[categorical_columns] = imputer_categorical.fit_transform(df[categorical_columns])

# Check if there are any remaining missing values
missing_values_after_imputation = df.isnull().sum()
print("Missing Values After Imputation:\n", missing_values_after_imputation)

# Display the updated dataset
df.head()
```

```python
if 'top_rated' in df.columns:
    df['top_rated'] = df['top_rated'].fillna(df['top_rated'].mode()[0])
else:
    print("'top_rated' column not found.")
```

```python
# Impute publish_date by converting it to datetime and use a mode imputation
df['publish_date'] = pd.to_datetime(df['publish_date'], errors='coerce')
df['publish_date'] = df['publish_date'].fillna(df['publish_date'].mode()[0])

# Impute 'top_rated' with the mode (most frequent value)
#df['top_rated'] = df['top_rated'].fillna('No')

# Check if there are any remaining missing values
missing_values_after_imputation = df.isnull().sum()
print("Missing Values After Imputation:\n", missing_values_after_imputation)

# Display the updated dataset
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize numerical columns
scaler = StandardScaler()
df[['rating', 'review_count', 'price','edition']] = scaler.fit_transform(df[['rating', 'review_count', 'price','edition']])

# Display the normalized data
print("\nNormalized Numerical Columns:\n", df[['rating', 'review_count', 'price','edition']].head())
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Encode categorical columns with LabelEncoder (or OneHotEncoder)
label_encoder = LabelEncoder()
df['best_seller'] = label_encoder.fit_transform(df['best_seller'])

# For the remaining categorical columns, apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=['title', 'description', 'author', 'isbn10', 'isbn13', 'publish_date'])

# Display the transformed DataFrame
print("\nTransformed DataFrame:\n", df_encoded)
```
